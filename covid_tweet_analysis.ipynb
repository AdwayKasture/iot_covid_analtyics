```python
import csv

file_name = "covid_data.csv"
rows = []
with open(file_name, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)
        
```

```python
for i in range(len(rows)):
    if rows[i][10] != 'Tweet':
        print(rows.pop(i))
    
```

```python
fields1 = ["city","age","date","oxygen","plasma","icu_beds","beds","ventilator","remdisivir"]
with open("final_covid.csv", 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(fields1)
    
    for row in rows :
        p = 0
        oxygen = 0
        plasma = 0
        ventilator = 0
        remdisvir = 0
        beds = 0
        icu_bed = 0
        age = 0
        city = ""
        datee = time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(row[5],'%a %b %d %H:%M:%S +0000 %Y'))
        try :
            p = 1
            #age = row[1].split("\n")[2].split("-")[1][1:]
            p = 2
            #city = row[1].split("\n")[3].split(". ")[-1]
            p = 3
            t1 = row[1].split()
            p = 4
            if("Oxygen" in row[1]):
                oxygen = 1
            p = 5
            if("Plasma" in row[1]):
                plasma = 1
            p = 6
            if("ICU" in row[1]):
                icu_bed = 1
            else:
                if("Bed" in row[1]):
                    beds = 1
            p = 7
            if("Ventilator" in row[1]):
                ventilator = 1
            p = 8
            if("emdisvir" in row[1]):
                remdisvir = 1
            p = 9
            l1 = [city,age,datee,oxygen,plasma,icu_bed,beds,ventilator,remdisvir]
            print(l1)
            csvwriter.writerow(l1)
        except :
            print(p)
```
